```python
import pandas as pd
import numpy as np

url = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis.csv"
df = pd.read_csv(url)

df.head()
```

```python
df.info()
df.describe(include="all")
```

```python
df.columns = df.columns.str.lower().str.replace(' ', '_')
df.head()
```

```python
df.isnull().sum()
```

```python
# Delete rows with too many null values
df = df.dropna(subset=['customer_lifetime_value', 'total_claim_amount'])

# fill null values in categorical columns with "Unknown"
df['state'] = df['state'].fillna("Unknown")
df['gender'] = df['gender'].fillna("Unknown")
```

```python
#Convert numerical values in (float/int)
df['customer_lifetime_value'] = df['customer_lifetime_value'].astype(float)
df['total_claim_amount'] = df['total_claim_amount'].astype(float)
df['monthly_premium_auto'] = df['monthly_premium_auto'].astype(float)
#Convert categorical values in string
df['gender'] = df['gender'].astype('string')
df['policy_type'] = df['policy_type'].astype('string')
df['education'] = df['education'].astype('string')
```

```python
#delete duplicates
df = df.drop_duplicates()
```

```python
#Reset Index
df = df.reset_index(drop=True)
```

```python
df.head()
```

```python
# 1 Display a new DataFrame
# total_claim_amount < 1000
# response == "Yes"

df_low_claim_yes = df[(df['total_claim_amount'] < 1000) & (df['response'] == 'Yes')]

print("Shape:", df_low_claim_yes.shape)

df_low_claim_yes.head()
```

```python
# 2 average monthly_premium and/or customer lifetime value by policy_type and gender for customers who responded "Yes"
df_yes = df[df['response'] == 'Yes']

agg_results = df_yes.groupby(['policy_type', 'gender']).agg({
    'monthly_premium_auto': 'mean',
    'customer_lifetime_value': 'mean',
    'total_claim_amount': 'mean'}).reset_index()

agg_results
```

```python
# 3 total number of customers in each state, and  only include states where there are more than 500 customers.

state_counts = df['state'].value_counts()
big_states = state_counts[state_counts > 500]

big_states


```

```python
# Find the maximum, minimum, and median customer lifetime value by education level and gender
clv_stats = df.groupby(['education', 'gender'])['customer_lifetime_value'].agg(['max', 'min', 'median']).reset_index()

clv_stats
```

# Conclusions

Median CLV is similar across genders (~5.3k–6.2k).

“High School or Below” shows the highest max CLV (~83k).

Doctoral customers have lower max CLV compared to other groups.

Education level seems more influential on CLV than gender.

```python
# 5 number of policies sold by state and month.
# Rows = state
# Columns = month
# Values = quantity policy_number (or clients)
# Make sure that 'effective_to_date' is Date type
df['effective_to_date'] = pd.to_datetime(df['effective_to_date'])

# Take the month
df['month'] = df['effective_to_date'].dt.month

# Create the DF
policies_state_month = pd.crosstab(df['state'], df['month'])

policies_state_month

```

```python
# 6. number of policies sold by month, by state, for the top 3 states
# Count total of policies por state
state_counts = df['state'].value_counts().head(3)

# Filter dataset 
top3_states = df[df['state'].isin(state_counts.index)]

# Count policies sold per state and month
top3_policies = top3_states.groupby(['state','month']).size().unstack()

top3_policies
```

```python
# 7 effect of different marketing channels on the customer response rate.
# Calculate answers per channel
channel_response = df.groupby(['sales_channel','response']).size().unstack(fill_value=0)

# Calculte rate of "Yes"
channel_response['response_rate_yes'] = channel_response['Yes'] / channel_response.sum(axis=1)

channel_response
```
